```python
nave = [
    {"nume": "Nava_A", "loa": 180, "teu": 1200},
    {"nume": "Nava_B", "loa": 300, "teu": 7500},
    {"nume": "Nava_C", "loa": 220, "teu": 2800},
    {"nume": "Nava_D", "loa": 150, "teu": 900},
]

# Selecteaza navele "mari" (teu >= 2000)
nave_mari = [n for n in nave if n["teu"] >= 2000]

print("Nave mari:")
for n in nave_mari:
    print("-", n["nume"], "| TEU:", n["teu"], "| LOA:", n["loa"])
```

```python
nave = [
    {"nume": "Nava_A", "loa": 180, "teu": 1200},
    {"nume": "Nava_B", "loa": 300, "teu": 7500},
    {"nume": "Nava_C", "loa": 220, "teu": 2800},
    {"nume": "Nava_D", "loa": 150, "teu": 900},
    {"nume": "Nava_E", "loa": 260, "teu": 5000},
]
```

```python
for n in nave:
    n["eficienta"] = n["teu"] / n["loa"]

for n in nave:
    print(n["nume"], "→ eficienta:", round(n["eficienta"], 2))

```

```python
nava_optima = max(nave, key=lambda x: x["eficienta"])

print("\nNava optimă este:")
print("Nume:", nava_optima["nume"])
print("TEU:", nava_optima["teu"])
print("LOA:", nava_optima["loa"])
print("Eficiență:", round(nava_optima["eficienta"], 2))
```
